```python
url='https://ceoworld.biz/2024/06/03/revealed-tallest-skyscrapers-in-the-world-in-2024/'
```

```python
pro=requests.get(url)
```

```python
soup=BeautifulSoup(pro.text,'html.parser')
```

```python
tabpro=soup.find('table')
print(tabpro)
```

```python
tabpro_th=tabpro.find_all('th')
```

```python
table_text=[th.text.strip() for th in tabpro_th]
table_text
```

```python
import pandas as pd
df=pd.DataFrame(columns=table_text)
df
```

```python
pro_column=tabpro.find_all('tr')
pro_table_tr=tabpro.find_all('tr')
pro_table_tr
```

```python
for table_td in pro_table_tr[1:]:
  table_td1=table_td.find_all('td')
  table_data1=[hr.text.strip() for hr in table_td1 ]
  length=len(df)
  df.loc[length]=table_data1

df
```

```python
from google.colab import files
```

```python
csv_file = 'Building.csv'
df.to_csv(csv_file, index=False)
files.download(csv_file)
```
